```python
import pandas as pd
import numpy as np
import datetime
```

```python
import os
destdir = './'

files = [ f for f in os.listdir(destdir) if os.path.isfile(os.path.join(destdir,f)) ]

```

```python
#Load the csv files

files[0]

df_2014 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2014__August_2013___June_2014_.csv',nrows=100000 ,low_memory=False)
df_2014=df_2014.dropna(axis=1, how='all')

df_2015 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2015.csv',nrows=100000,low_memory=False)
df_2015=df_2015.dropna(axis=1, how='all')

df_2016= pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2016.csv',nrows=100000,low_memory=False)
df_2016=df_2016.dropna(axis=1, how='all')

df_2017 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2017.csv',nrows=100000,low_memory=False)
df_2017=df_2017.dropna(axis=1, how='all')

# df_2014.head()
# df_2014.columns
```

```python
df_2014.shape
```

```python
#Creation and clean-up of dataframes.Few unwanted columns were dropped as well as NAN values were removed
#The data was filtered for some specific values,like Registration state='NY',Vehical type='Passanger' and vehicle body type='4 door sedan'
# 2014
df2014 = pd.DataFrame(df_2014,columns=['Summons Number', 'Plate ID', 'Registration State', 'Plate Type',
       'Issue Date', 'Violation Code', 'Vehicle Body Type', 'Vehicle Make',
       'Issuing Agency','Issuer Precinct','Violation County', 'Vehicle Year', 'Violation Post Code', 'Violation Description'])
df2014=df2014[df2014['Registration State']=='NY']
df2014=df2014[df2014['Plate Type']=='PAS'] 
df2014=df2014[ df2014['Vehicle Body Type']=='4DSD']
df2014.reset_index(inplace=True,drop=True)
df2014.head()

#2015
df2015 = pd.DataFrame(df_2015,columns=['Summons Number', 'Plate ID', 'Registration State', 'Plate Type',
       'Issue Date', 'Violation Code', 'Vehicle Body Type', 'Vehicle Make',
       'Issuing Agency','Issuer Precinct','Violation County', 'Vehicle Year', 'Violation Post Code', 'Violation Description'])
df2015=df2015[df2015['Registration State']=='NY']
df2015=df2015[df2015['Plate Type']=='PAS'] 
df2015=df2015[ df2015['Vehicle Body Type']=='4DSD']
df2015.reset_index(inplace=True,drop=True)
df2015.head()

#2016
df2016 = pd.DataFrame(df_2016,columns=['Summons Number', 'Plate ID', 'Registration State', 'Plate Type',
       'Issue Date', 'Violation Code', 'Vehicle Body Type', 'Vehicle Make',
       'Issuing Agency','Issuer Precinct','Violation County', 'Vehicle Year', 'Violation Post Code', 'Violation Description'])
df2016=df2016[df2016['Registration State']=='NY']
df2016=df2016[df2016['Plate Type']=='PAS'] 
df2016=df2016[ df2016['Vehicle Body Type']=='4DSD']
df2016.reset_index(inplace=True,drop=True)
df2016.head()


#2017
df2017 = pd.DataFrame(df_2017,columns=['Summons Number', 'Plate ID', 'Registration State', 'Plate Type',
       'Issue Date', 'Violation Code', 'Vehicle Body Type', 'Vehicle Make',
       'Issuing Agency','Issuer Precinct','Violation County', 'Vehicle Year', 'Violation Post Code', 'Violation Description'])
df2017=df2017[df2017['Registration State']=='NY']
df2017=df2017[df2017['Plate Type']=='PAS'] 
df2017=df2017[ df2017['Vehicle Body Type']=='4DSD']
df2017.reset_index(inplace=True,drop=True)
df2017.head()

```

```python
df2014.shape
df2015.shape
df2016.shape
df2017.shape
```

```python
#the data for 4 years is concatenated/appended in a single dataframe
parking_df=pd.concat([df2014, df2015, df2016, df2017], axis=0)
```

```python
#the final data frame is cleaned for NAN values
parking_df.dropna()
parking_df.tail()
parking_df.shape
```
