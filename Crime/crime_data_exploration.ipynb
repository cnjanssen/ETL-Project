```python
import pandas as pd
import numpy as np
import datetime
```

```python
import os
destdir = './'

files = [ f for f in os.listdir(destdir) if os.path.isfile(os.path.join(destdir,f)) ]
```

```python
files
```

```python
df_mis_off=pd.read_csv('misdemeanor-offenses-by-precinct-2000-2018.csv',nrows=100000,low_memory=False).dropna(axis=1, how='all')
df_non_maj=pd.read_csv('non-seven-major-felony-offenses-by-precinct-2000-2018.csv',nrows=100000,low_memory=False).dropna(axis=1, how='all')
df_maj_fel=pd.read_csv('seven-major-felony-offenses-by-precinct-2000-2018.csv',nrows=100000,low_memory=False).dropna(axis=1, how='all')
df_vio_off=pd.read_csv('violation-offenses-by-precinct-2000-2018.csv',nrows=100000,low_memory=False).dropna(axis=1, how='all')
```

```python
df_vio_off.head()
```

```python
crime_df=pd.concat([df_mis_off, df_non_maj, df_maj_fel, df_vio_off], axis=0)
```

```python

crime_df.sort_values(by=['PCT'], inplace=True)
```

```python
ind_drop = crime_df[crime_df['CRIME'].apply(lambda x: x.startswith('TOTAL'))].index
```

```python
new_crime_df = crime_df.drop(ind_drop).reset_index(drop=True)
```
